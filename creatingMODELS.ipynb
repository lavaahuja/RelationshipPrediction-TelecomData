```python
from tensorflow._api.v2.compat.v1 import keras
from keras.regularizers import l2
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold, train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.utils import class_weight
from keras.models import Sequential
from keras.layers import Dense, Dropout
import matplotlib.pyplot as plt

# Load the CSV file
input_file_path = 'hawkes-networksurveyCOMPLETE.csv'
df = pd.read_csv(input_file_path, sep=';')
df1 = pd.read_csv('RESULTS/COMPLIED_POWER_LAW_RESULTS.csv')
df1 = df1.replace([np.inf, -np.inf], np.nan)
df1 = df1.dropna()

df2 = pd.read_csv('RESULTS/CompiledARIMAResultFits_all.csv')

# Perform inner join on 'EgoID' and 'AlterID'
result_df = pd.merge(df, df1, left_on=['EgoID', 'AlterID'], right_on=['egoid', 'alterid'], how='inner')

# Drop the additional columns from the result
result_df = result_df.drop(['egoid', 'alterid'], axis=1)

result_df = pd.merge(result_df, df2, left_on=['EgoID', 'AlterID'], right_on=['egoid', 'alterid'], how='inner')
result_df = result_df.drop(['egoid', 'alterid'], axis=1)

# Display the result
result_df = result_df.sort_values(by='EgoID').reset_index(drop=True)


friendship_relations = ['friendship-stable', 'friendship-relaxing', 'friendship-fluctuating', 'colleagues-stable',
                        'friendship-strengthening', 'colleagues-relaxing']
family_relations = ['family-stable', 'family-relaxing', 'other-stable', 'other-relaxing']
romantic_relations = ['romantic-relaxing', 'romantic-fluctuating', 'romantic-strengthening', 'romantic-stable']

result_df['Types'] = np.where(result_df['RelationshipOverTime'].isin(friendship_relations), 'friendship_relations',
                              np.where(result_df['RelationshipOverTime'].isin(family_relations), 'family_relations',
                                       np.where(result_df['RelationshipOverTime'].isin(romantic_relations),
                                                'romantic_relations', None)))

result_df = result_df.dropna(subset=['Types'])

stable_categories = ['friendship-stable', 'family-stable', 'romantic-stable', 'colleagues-stable', 'other-stable']
stren_categories = ['friendship-strengthening', 'romantic-strengthening']
relaxing_categories = ['friendship-relaxing', 'family-relaxing', 'romantic-relaxing', 'other-relaxing',
                       'colleagues-relaxing', 'romantic-fluctuating', 'friendship-fluctuating']
varied_categories = ['unknown', 'rnd']

result_df['Category'] = np.where(result_df['RelationshipOverTime'].isin(stable_categories),
                                 'Stable Relationships',
                                 np.where(result_df['RelationshipOverTime'].isin(relaxing_categories),
                                          'Relaxing Relationships',
                                          np.where(result_df['RelationshipOverTime'].isin(stren_categories),
                                                   'Strengthening Relationships',
                                                   np.where(result_df['RelationshipOverTime'].isin(varied_categories),
                                                            'Varied Relationships', 'None'))))
result_df = result_df.dropna(subset=['Category'])

# Select features and labels
features = ['mu', 'k', 'p_x', 'c', 'branching factor', 'p_y', 'd', 'q']
label_column = 'Types'

# Create a mapping between original class names and their corresponding encoded values
class_mapping = dict(enumerate(result_df[label_column].unique()))

# Create a reverse mapping for decoding
reverse_class_mapping = {v: k for k, v in class_mapping.items()}

# Apply the mapping to create a new column with encoded labels
result_df['Encoded_Labels'] = result_df[label_column].map(reverse_class_mapping)

# Split the data into training and testing sets
train_data, test_data, train_labels, test_labels = train_test_split(
    result_df[features], result_df['Encoded_Labels'], test_size=0.1, random_state=62, stratify=result_df['Encoded_Labels']
)

# Standardize the features
scaler = StandardScaler()
train_data_scaled = scaler.fit_transform(train_data)
test_data_scaled = scaler.transform(test_data)

# Apply SMOTE to oversample the minority class
smote = SMOTE(random_state=62)
train_data_resampled, train_labels_resampled = smote.fit_resample(train_data_scaled, train_labels)

class_counts = pd.Series(train_labels_resampled).value_counts()
print("Class Counts Post-SMOTE:")
print(class_counts)

# Define the number of folds for cross-validation
num_folds = 5
kfold = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=62)

# Initialize lists to store performance metrics for each fold
accuracy_list = []
precision_list = []
recall_list = []
f1_list = []

for train_index, val_index in kfold.split(train_data_resampled, train_labels_resampled):
    # Split the data into training and validation sets for each fold
    kfold_train_data, kfold_val_data = train_data_resampled[train_index], train_data_resampled[val_index]
    kfold_train_labels, kfold_val_labels = train_labels_resampled[train_index], train_labels_resampled[val_index]

    # Create and compile the model
    model = Sequential()
    model.add(Dense(256, input_dim=len(features), activation='relu', kernel_regularizer=l2(0.01)))
    model.add(Dropout(0.5))
    model.add(Dense(128, activation='relu', kernel_regularizer=l2(0.01)))
    model.add(Dropout(0.5))
    model.add(Dense(len(class_mapping), activation='softmax'))

    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

    # Train the model on the current fold
    model.fit(kfold_train_data, kfold_train_labels, epochs=50, batch_size=32, validation_data=(kfold_val_data, kfold_val_labels), validation_split=0.2,callbacks=[early_stopping], verbose=1)

    # Make predictions on the validation set
    kfold_val_predictions = np.argmax(model.predict(kfold_val_data), axis=1)

    # Evaluate the model on the validation set
    val_accuracy = accuracy_score(kfold_val_labels, kfold_val_predictions)
    val_precision = precision_score(kfold_val_labels, kfold_val_predictions, average='weighted')
    val_recall = recall_score(kfold_val_labels, kfold_val_predictions, average='weighted')
    val_f1 = f1_score(kfold_val_labels, kfold_val_predictions, average='weighted')

    # Append the metrics to the lists
    accuracy_list.append(val_accuracy)
    precision_list.append(val_precision)
    recall_list.append(val_recall)
    f1_list.append(val_f1)

# Calculate and print the mean metrics across all folds
mean_accuracy = np.mean(accuracy_list)
mean_precision = np.mean(precision_list)
mean_recall = np.mean(recall_list)
mean_f1 = np.mean(f1_list)

print(f'Mean Accuracy: {mean_accuracy}')
print(f'Mean Precision: {mean_precision}')
print(f'Mean Recall: {mean_recall}')
print(f'Mean F1-Score: {mean_f1}')

# Make predictions on the test set
nn_predictions = np.argmax(model.predict(test_data_scaled), axis=1)

# Decode the predictions back to original labels
decoded_nn_predictions = pd.Series(nn_predictions).map({i: label for i, label in enumerate(class_mapping)})

# Evaluate the Neural Network model on the test set
nn_accuracy = accuracy_score(test_labels, decoded_nn_predictions)
nn_precision = precision_score(test_labels, decoded_nn_predictions, average='weighted')
nn_recall = recall_score(test_labels, decoded_nn_predictions, average='weighted')
nn_f1 = f1_score(test_labels, decoded_nn_predictions, average='weighted')
nn_conf_matrix = confusion_matrix(test_labels, decoded_nn_predictions)

# Print metrics for Neural Network on the test set
print(f'Neural Network Test Accuracy: {nn_accuracy}')
print(f'Neural Network Test Precision: {nn_precision}')
print(f'Neural Network Test Recall: {nn_recall}')
print(f'Neural Network Test F1-Score: {nn_f1}')
print('Neural Network Test Confusion Matrix:')
print(nn_conf_matrix)

# Plot ROC curve for each class on the test set
fpr = dict()
tpr = dict()
roc_auc = dict()

# Binarize the labels for multi-class ROC
test_labels_bin = label_binarize(test_labels, classes=np.unique(decoded_nn_predictions))

for i in range(len(class_mapping)):
    # Create binary labels for the current class
    y_true = np.where(test_labels == i, 1, 0)
    # Extract the probability for the positive class
    y_score = model.predict(test_data_scaled)[:, i]

    # Compute ROC curve and AUC
    fpr[i], tpr[i], _ = roc_curve(y_true, y_score)
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot ROC curve for each class on the test set
plt.figure(figsize=(8, 6))
for i in range(len(class_mapping)):
    plt.plot(fpr[i], tpr[i], label=f'Class {class_mapping[i]} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve for Multi-Class on Test Set')
plt.legend(loc='lower right')
plt.show()
```
