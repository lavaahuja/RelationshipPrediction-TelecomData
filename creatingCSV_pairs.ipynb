```python
import pyodbc
import pandas as pd

# Set up the connection string
server = 'localhost'
database = 'telecomDB'
conn_str = f'DRIVER={{SQL Server}};SERVER={server};DATABASE={database};Trusted_Connection=yes;'

# Connect to the database
conn = pyodbc.connect(conn_str)
cursor = conn.cursor()
```

```python
import pandas as pd
# Read the original CSV file
df = pd.read_csv('data/CommunicationEventsSELECTIVE.csv')
i=1
```

```python
df.columns
```

```python
i=1
# Iterate through unique combinations of egoid and alterid
for (egoid, alterid), group in df.groupby(['egoid', 'alterid']):
    # Create a new DataFrame with columns: egoid, alterid, timestamp, outgoing
    new_df = pd.DataFrame({
        'egoid': [egoid] * len(group),
        'alterid': [alterid] * len(group),
        'timestamp': group['timestamp'],
        'outgoing': (group['outgoing'] == 'Yes').astype(int)  # Convert 'Yes' to 1, 'No' to 0
    })

    # Save the new DataFrame to a CSV file
    file_name = f'{egoid}_{alterid}.csv'
    new_df.to_csv(file_name, index=False)
    print(f'Processed {egoid}_{alterid}. {i} file(s) done.')
    i += 1
```
